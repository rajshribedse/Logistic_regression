```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Disable Warnings
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter("ignore")
```

```python
data = pd.read_csv("/content/payment_fraud_2.csv")
```

```python
data.paymentMethod = data.paymentMethod.map({"paypal":"PayPal",
                                            "storecredit": "Store Credit", #A store credit is a document offered by a store to a customer who returns an item not eligible for a refund.
                                            "creditcard": "Credit Card"})
```

```python
data.paymentMethod.unique()
```

```python
payment_counts = data['paymentMethod'].value_counts()
```

```python
# Create a pie chart
fig = px.pie(names=payment_counts.index, values=payment_counts.values, title='Payment Method Distribution')

# Show the pie chart
fig.show()
```

```python
x = np.array(data[["accountAgeDays", "numItems", "localTime", "paymentMethodAgeDays"]])
y = np.array(data[["label"]])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,
                                               y,
                                               test_size = 0.2,
                                               random_state = 42)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(xtrain, ytrain)
logistic_model.score(xtrain, ytrain)
```

```python
yhat = logistic_model.predict(xtest)
```

```python
print(f"Accuracy Score of Logistic Regression Model: {accuracy_score(yhat, ytest)}")
```

```python
print(f"Confusion Matrix of Logistic Regression Model:\n {confusion_matrix(yhat, ytest)}")
```
